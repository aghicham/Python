```python
sns.countplot(data=df, x="Survived", hue="Sex", palette=['r', 'b'])
```

```python
ax = sns.barplot(data=df, x="Survived", y="Age", hue="Sex", palette=['b', 'magenta'])
ax.grid()
```

```python
sns.histplot(data=df, x="Age", hue="Sex", palette=['b', 'magenta'], multiple="stack", bins=8)
```

```python
sns.histplot(data=df, x="Age", hue="Sex", palette=['b', 'magenta'], multiple="stack", binwidth=5)
```

```python
sns.kdeplot(data=df, x="Age", hue="Sex", palette=['b', 'magenta'])
```

```python
sns.histplot(data=df, x="Age", hue="Sex", palette=['b', 'magenta'], multiple="stack", binwidth=10, kde=True)
```

```python
g = sns.FacetGrid(data=df, col="Survived")
g.map(sns.barplot, "Sex", "Age", "Pclass", palette=["b", "r", "g"])
```

```python
sns.displot(data=df, x="Age", hue="Sex", palette=['b', 'magenta'], multiple="stack", binwidth=10, kde=True, col="Survived")
```

```python
sns.displot(data=df, x="Age", hue="Sex", palette=['b', 'magenta'], row="Survived", col="Pclass", multiple="stack")
```
