```python
import tkinter as tk
import random

# Initialize the game window
window = tk.Tk()
window.title("Rock Paper Scissors Game")
window.geometry("400x300")
window.configure(bg="lightblue")

# Create a label for the game title
title_label = tk.Label(window, text="Rock Paper Scissors", font=("Helvetica", 18, "bold"), bg="lightblue")
title_label.pack(pady=10)

# Prompt the user to choose between rock, paper, or scissors
prompt_label = tk.Label(window, text="Choose Rock, Paper, or Scissors:", font=("Helvetica", 12), bg="lightblue")
prompt_label.pack(pady=5)

# Create an input field for the user to enter their choice
player_choice = tk.StringVar(value=None)
choice_entry = tk.Entry(window, textvariable=player_choice, font=("Helvetica", 12))
choice_entry.pack(pady=5)

# Generate a random selection for the computer
choices = ["Rock", "Paper", "Scissors"]
comp_pick = random.choice(choices)

# Function to handle game logic and determine the winner
def play():
    user_choice = player_choice.get()
    comp_choice = random.choice(choices)
    
    if user_choice == comp_choice:
        result = "It's a tie!"
    elif (user_choice == "Rock" and comp_choice == "Scissors") or \
         (user_choice == "Scissors" and comp_choice == "Paper") or \
         (user_choice == "Paper" and comp_choice == "Rock"):
        result = "You win!"
    else:
        result = "Computer wins!"
    
    result_label.config(text=f"Computer chose: {comp_choice}\n{result}")

# Function to reset the game
def reset():
    player_choice.set("")
    result_label.config(text="")

# Function to exit the application
def exit_game():
    window.destroy()

# Create a button to play the game
play_button = tk.Button(window, text="Play", command=play)
play_button.pack(pady=5)

# Create buttons to reset the game and exit the application
reset_button = tk.Button(window, text="Reset", command=reset)
reset_button.pack(pady=5)

exit_button = tk.Button(window, text="Exit", command=exit_game)
exit_button.pack(pady=5)

# Label to display the result of the game
result_label = tk.Label(window, text="", font=("Helvetica", 12), bg="lightblue")
result_label.pack()

# Start the main loop of the application
window.mainloop()
```
